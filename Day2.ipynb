```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', data=df, s=100)
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))

# Get unique species names and map them to numerical labels for coloring
species_names = df['species'].unique()
for i, species_name in enumerate(species_names):
    subset = df[df['species'] == species_name]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'], label=species_name)

plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```
