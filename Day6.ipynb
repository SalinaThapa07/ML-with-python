```python
#Pipeline function
#it connect a model with its necessary preprocessing steps,allowing us to directly input any text and get an intelligible answer
```

```python
from transformers import pipeline
```

```python
classifier= pipeline("sentiment-analysis")
classifier("I have been waiting for a HuggingFace course my whole life.")
```

```python
classifier= pipeline("sentiment-analysis")
classifier(["I dont like exam.", "I hate reading so much"])
```

```python
classifier =pipeline("zero-shot-classification")
classifier(
    " football",
    candidate_labels=["education", "politics", "business","Messi"],
)
```

```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

```python
ner= pipeline("ner", grouped_entities=True)
ner("My name is Salina and I work at Hugging Face in Brooklyn.")
```
